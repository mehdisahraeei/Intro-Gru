### Importing

```python
from numpy import array
from numpy import hstack
from keras.models import  Sequential
from keras.layers import GRU, Dropout, Dense, TimeDistributed, RepeatVector
```

### method

```python
def split_sequence(sequence, n_step_in, n_step_out):
   x,y = list(), list() #x is data and y is label
   for i in range(len(sequence)):
       end_ix = i + n_step_in #timestep
       out_end_ix = end_ix + n_step_out # n-output
       if out_end_ix > len(sequence): #In this -if- we want if data be beyond len(sequence) , the command will not continue and break.
          break
       seq_x, seq_y = sequence[i:end_ix] , sequence[end_ix:out_end_ix]   
       x.append(seq_x)
       y.append(seq_y)
   return array(x) , array(y)
```

```python
n_step_in, n_step_out = 3 , 2   
seq = [10,20,35,50,55,64,70,80,95]
      
x,y = split_sequence(seq, n_step_in, n_step_out)
print(x.shape,y.shape)   
   
for i in range(len(x)):
   print(x[i],y[i])
```

```python
x.shape
```

### training model

```python
#The input must be three-dimensional or three-channel, that's why we reshape it
n_features = 1
x = x.reshape(x.shape[0],x.shape[1],n_features)    
y = y.reshape(y.shape[0],y.shape[1],n_features)
```

```python
model = Sequential()

model.add(GRU(units=200, activation='tanh',  input_shape=(n_step_in,n_features))) # tanh is best activation in gru but relu also is used
model.add(RepeatVector(n_step_out))   
model.add(GRU(150,activation = 'tanh',return_sequences=True))   
model.add(TimeDistributed(Dense(1)))
model.compile(optimizer='RMSprop',loss='mse')   
```

```python
model.fit(x,y,epochs=300,verbose=1)
```

### testing

```python
x_input = array([10,20,35])
x_input = x_input.reshape(1, n_step_in, n_features)
x_input.shape
```

```python
yhat = model.predict(x_input,verbose=0)
print('hat y:',yhat,'\n')
print('real y',y[0])
```
